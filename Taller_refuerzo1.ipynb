1.Un vendedor recibe un sueldo base más un 10% extra por
comisión de sus ventas, el vendedor desea saber cuánto
dinero obtendrá por concepto de comisiones por las tres
ventas que realiza en el mes y el total que recibirá en
el mes tomando en cuenta su sueldo base y comisiones.

```python
sueldob = float(input("ingrese su sueldo base "))
venta1 = float(input("Ingrese el valor de la venta 1 "))
venta2 = float(input("Ingrese el valor de la venta 2 "))
venta3 = float(input("Ingrese el valor de la venta 3 "))

comision = (venta1 + venta2 + venta3) * 0.10
sueldoc = (sueldob + comision)

print(f"\nEl sueldo que se ingreso es {sueldob:,.0f}")
print(f"\nla comision por las ventas de este mes es {comision:,.0f}")
print(f"\nEl sueldo mas comision es {sueldoc:,.0f}")
```

2. Una tienda ofrece un descuento del 15% sobre el total
de la compra y un cliente desea saber cuánto deberá
pagar finalmente por su compra.

```python
compra = float(input("ingrese el valor de la compra "))

descuento = round(compra * 0.15,2)
total = compra  - descuento 

print(f"El descuento aplicado es {descuento:,.0f}%")
print(f"el valor total a pagar es {total:,.0f}")
```

3. Un alumno desea saber cuál será su calificación final en
la materia de Algoritmos. Dicha calificación se compone
de tres exámenes parciales.

```python
cal1= float(input("digite la calificacion 1: "))
cal2= float(input("digite la calificacion 2: "))
cal3= float(input("digite la calificacion 3: "))

promedio= round((cal1 + cal2 + cal3)/3, 2) # round funcion de redondear

print(f"el promedio es ={promedio}")
```

4. Un maestro desea saber qué porcentaje de hombres y que
porcentaje de mujeres hay en un grupo de estudiantes.

```python
hombres = int(input("escriba la cantidad de estudiantes hombres "))
mujeres = int(input("escriba la cantidad de estudiantes mujeres "))

total = mujeres + hombres 
po_hombres = round((hombres/total)*100,2) 
po_mujeres = round((mujeres/total)*100,2) 

print(f"el porcentaje de estudiantes hombres es {po_hombres}% ")
print(f"el porcentaje de estudiantes mujeres es {po_mujeres}% ")
```
